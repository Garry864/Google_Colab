```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data
sample_data = np.array([5.1, 5.5, 4.9, 5.3, 5.2, 5.6, 5.0, 5.7, 5.4, 5.3])

# Population mean (specified value)
population_mean = 5.0

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Degrees of freedom
df = len(sample_data) - 1

# Critical t-value (two-tailed test)
alpha = 0.05
critical_t_value = stats.t.ppf(1 - alpha/2, df)

# Compare test statistic and critical region
if np.abs(t_statistic) > critical_t_value:
    reject_null_hypothesis = True
else:
    reject_null_hypothesis = False

# Plot the sample distribution and critical region
plt.figure(figsize=(10, 6))

# Histogram of sample data
plt.hist(sample_data, bins=10, alpha=0.7, color='blue', edgecolor='black', label='Sample Data')

# Critical region boundaries
plt.axvline(x=population_mean - critical_t_value*np.std(sample_data)/np.sqrt(len(sample_data)), color='red', linestyle='--', label='Critical Region')
plt.axvline(x=population_mean + critical_t_value*np.std(sample_data)/np.sqrt(len(sample_data)), color='red', linestyle='--')

# Sample mean
plt.axvline(x=np.mean(sample_data), color='green', linestyle='-', label='Sample Mean')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('One-Sample t-Test')
plt.legend()
plt.grid(True)

plt.show()

# Interpret results
if reject_null_hypothesis:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")

if p_value < alpha:
    print("The population mean is significantly different from the specified value.")
else:
    print("There is not enough evidence to conclude that the population mean is different from the specified value.")
```

Q1)According to a study, the daily average time spent by a user on a
social media website is 50 minutes. To test the claim of this study,
Ramesh, a researcher, takes a sample of 25 website users and finds
out that the mean time spent by the sample users is 60 minutes and
the sample standard deviation is 30 minutes. <br>
Based on this information, the null and the alternative hypotheses will
be:<br>

a. Ho = The average time spent by the users is 50 minutes <br>
b. H1 = The average time spent by the users is not 50
minutes <br>

Use a 5% significance level to test this hypothesis.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
sample_mean = 60  # Sample mean
sample_std_dev = 30  # Sample standard deviation
n = 25  # Sample size
population_mean = 50  # Population mean
alpha = 0.05  # Significance level

# Calculate the test statistic (z-score)
z_score = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(n))

# Calculate the critical values
critical_value_left = stats.norm.ppf(alpha / 2)
critical_value_right = stats.norm.ppf(1 - alpha / 2)

# Visualize the distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(critical_value_left, color='red', linestyle='--', label='Critical Value (Left)')
plt.axvline(critical_value_right, color='red', linestyle='--', label='Critical Value (Right)')
plt.axvline(z_score, color='green', linestyle='-', label='Test Statistic (z-score)')
plt.fill_between(x, 0, y, where=(x <= critical_value_left) | (x >= critical_value_right), color='red', alpha=0.3)
plt.title('Hypothesis Testing with Z-test')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')

# Adjust legend parameters
plt.legend(loc='upper left', fontsize='small')

plt.show()

# Perform the hypothesis test
if z_score < critical_value_left or z_score > critical_value_right:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Q5)For a certain type of computer, the length of time between
charges of the battery is normally distributed with a mean of 50 hours
and a standard deviation of 15 hours. John owns one of these
computers and wants to know the probability that the length of time
will be between 50 and 70 hours.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given data
mean_time = 50  # Mean time between charges (hours)
std_dev = 15  # Standard deviation (hours)

# Calculate the probability
x = np.linspace(0, 100, 1000)
pdf = stats.norm.pdf(x, mean_time, std_dev)
plt.plot(x, pdf, label='Normal Distribution')
plt.fill_between(x, pdf, where=(x >= 50) & (x <= 70), color='red', alpha=0.3)
plt.xlabel('Time Between Charges (hours)')
plt.ylabel('Probability Density')
plt.title('Probability of Time Between Charges')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the probability
prob_between_50_and_70 = stats.norm.cdf(70, loc=mean_time, scale=std_dev) - stats.norm.cdf(50, loc=mean_time, scale=std_dev)

# Conclusion
print("Probability that time between charges will be between 50 and 70 hours:", prob_between_50_and_70)
```

Q7)It is estimated that 50% of emails are spam emails. Some software
has been applied to filter these spam emails before they reach your
inbox. A certain brand of software claims that it can detect 99% of
spam emails, and the probability of a false positive (a non-spam email
detected as spam) is 5%. Now if an email is detected as spam, then
what is the probability that it is in fact a non-spam email?

```python
# Given probabilities
P_spam = 0.5  # Probability of a spam email
P_non_spam = 1 - P_spam  # Probability of a non-spam email

P_detected_spam_given_spam = 0.99  # Probability of detecting spam given it's actually spam
P_detected_spam_given_non_spam = 0.05  # Probability of detecting spam given it's actually non-spam

# Calculate the overall probability of an email being detected as spam
P_detected_spam = P_detected_spam_given_spam * P_spam + P_detected_spam_given_non_spam * P_non_spam

# Calculate the probability that an email detected as spam is actually a non-spam email using Bayes' theorem
P_non_spam_given_detected_spam = (P_detected_spam_given_non_spam * P_non_spam) / P_detected_spam

print("Probability that an email detected as spam is actually a non-spam email:", P_non_spam_given_detected_spam)
```

```python
# Given probabilities
p_spam = 0.5  # Probability of spam email
p_detect_spam = 0.99  # Probability of detecting spam email
p_false_positive = 0.05  # Probability of false positive

# Calculate the probability of non-spam email detection
p_non_spam_given_detected = (1 - p_spam) * p_false_positive / ((1 - p_spam) * p_false_positive + p_spam * (1 - p_detect_spam))

# Conclusion
print("Probability that a detected email is a non-spam email:", p_non_spam_given_detected)
```

Q8)Given the following distribution of returns, determine the lower
quartile: <br>
{10 25 12 21 19 17 16 11 15 19}

```python
# Given data
data = [10, 25, 12, 21, 19, 17, 16, 11, 15, 19]

# Step 1: Sort the data in ascending order
sorted_data = sorted(data)

# Step 2: Determine the position of the lower quartile
n = len(sorted_data)
position = 1/4 * (n + 1)

# Step 3: Interpolate if necessary
if position.is_integer():
    lower_quartile = sorted_data[int(position) - 1]  # -1 to account for 0-based indexing
else:
    lower_index = int(position)  # Integer part of the position
    fractional_part = position - lower_index  # Fractional part
    lower_quartile = sorted_data[lower_index - 1] + fractional_part * (sorted_data[lower_index] - sorted_data[lower_index - 1])

print("Lower quartile:", lower_quartile)
```

Q9)For a Binomial distribution, the number of trials(n) is 25, and the
probability of success is 0.3. Whats the variability of the distribution?

```python
# n = 25
# p = 0.3
```

```python
import numpy as np

# Given parameters
n = 25  # Number of trials
p = 0.3  # Probability of success

# Calculate the variance of the binomial distribution
variance = np.var(np. random.binomial(n, p, 10000))

print("Variability (Variance) of the binomial distribution:", variance)
```

```python
# Given parameters
n = 25  # Number of trials
p = 0.3  # Probability of success

# Calculate the variance of the binomial distribution
variance = n * p * (1 - p)

print("Variability (Variance) of the binomial distribution:", variance)
```

```python
# Given parameters
n_trials = 25  # Number of trials
p_success = 0.3  # Probability of success

# Calculate the variability (variance) of the binomial distribution
variability = n_trials * p_success * (1 - p_success)

# Conclusion
print("Variability of the binomial distribution:", variability)
```

```python
# Import necessary libraries
import pandas as pd
from scipy.stats import f_oneway

# Load the dataset
data = pd.read_csv("Cell Phone Survey.csv")

# Perform ANOVA test
f_statistic, p_value = f_oneway(data[data['Type'] == 'Smart']['Value for the Dollar'],
                                data[data['Type'] == 'Camera']['Value for the Dollar'],
                                data[data['Type'] == 'Basic']['Value for the Dollar'])

# Display ANOVA results
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

# Interpretation and Conclusion
if p_value < 0.05:
    print("Conclusion: There is a significant difference in the mean response for 'Value for the Dollar' among different types of cell phones.")
else:
    print("Conclusion: There is no significant difference in the mean response for 'Value for the Dollar' among different types of cell phones.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
from scipy.stats import f_oneway

# Load the dataset
df = pd.read_csv("Cell Phone Survey.csv")

# 1. Checking datatypes of each column
print("Datatypes of each column:")
print(df.dtypes)

# 2. Find the Mean of the Signal strength column
mean_signal_strength = df["Signal strength"].mean()

# 3. Find the Median of Customer Service column
median_customer_service = df["Customer Service"].median()

# 4. Find the Mode of Signal strength column
mode_signal_strength = df["Signal strength"].mode()[0]

# 5. Find the Standard deviation of the Customer Service column
std_customer_service = df["Customer Service"].std()

# 6. Find the Variance of Customer Service column
var_customer_service = df["Customer Service"].var()

# 7. Calculate Percentiles of Value for the Dollar column
percentiles_value_dollar = np.percentile(df["Value for the Dollar"], [25, 50, 75])

# 8. Calculate the Range of Value for the Dollar column
range_value_dollar = df["Value for the Dollar"].max() - df["Value for the Dollar"].min()

# 9. Calculate IQR of Value for the Dollar column
q1 = df["Value for the Dollar"].quantile(0.25)
q3 = df["Value for the Dollar"].quantile(0.75)
iqr_value_dollar = q3 - q1

# 10. Hypothesis Testing - ANOVA
# Group data by 'Type' and extract 'Value for the Dollar' column
grouped_data = df.groupby('Type')['Value for the Dollar'].apply(list)

# Perform ANOVA test
f_statistic, p_value = f_oneway(*grouped_data)

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Value for the Dollar', data=df)
plt.title('Boxplot of Value for the Dollar by Cell Phone Type')
plt.xlabel('Cell Phone Type')
plt.ylabel('Value for the Dollar')
plt.show()

# Conclusion
print("\nMean of Signal strength column:", mean_signal_strength)
print("Median of Customer Service column:", median_customer_service)
print("Mode of Signal strength column:", mode_signal_strength)
print("Standard deviation of Customer Service column:", std_customer_service)
print("Variance of Customer Service column:", var_customer_service)
print("Percentiles of Value for the Dollar column:", percentiles_value_dollar)
print("Range of Value for the Dollar column:", range_value_dollar)
print("IQR of Value for the Dollar column:", iqr_value_dollar)
print("\nF-statistic for ANOVA test:", f_statistic)
print("p-value for ANOVA test:", p_value)
```
